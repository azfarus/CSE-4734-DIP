```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm


img = np.zeros((200,200))

img[50:100,50:100] = np.ones((50,50))
img[100:150,50:100] = np.ones((50,50))*0.25
img[50:100,100:150] = np.ones((50,50))*0.5
img[100:150,100:150] = np.ones((50,50))*0.1

# plt.imshow(img,cmap='gray')

img_fft = np.fft.fft2(img,norm='ortho')

ifft = np.fft.ifftshift(img_fft)
fft = np.fft.fftshift(img_fft)
 
plt.imshow((np.abs(img)),cmap='gray')
plt.show()
plt.imshow((np.abs(img_fft)),cmap='gray' , norm=LogNorm())
plt.show()
plt.imshow(np.abs(ifft),cmap='gray',norm=LogNorm())
plt.show()
plt.imshow(np.abs(fft),cmap='gray',norm=LogNorm())
plt.show()
plt.imshow(np.abs(np.fft.ifft2(np.fft.fftshift(ifft))),cmap='gray')
plt.show()
plt.imshow(np.abs(np.fft.ifft2(np.fft.fftshift(fft))),cmap='gray')
plt.show()
```
